```python
import nltk
import string
nltk.download('punkt')
```

```python
text = """
I usually like listening to music while studying. I believe that it helps me focus, while some people feel it can be distracting. The real question is: does music improve concentration, or does it reduce productivity?

Healthy habits are important to be followed by everyone. Eating nutritious food, exercising regularly, and getting enough sleep daily can improve us both physical and mental health. Doctors often say that: small lifestyle changes can make a big difference!
"""
print("Original Text:\n")
print(text)
```

```python
text_remove_punctuation = text.translate(str.maketrans('', '', string.punctuation))

print(text_remove_punctuation)
```

```python
nltk.download('stopwords')
nltk.download('punkt_tab')
nltk.download('wordnet')
from nltk.corpus import stopwords

words = nltk.word_tokenize(text_remove_punctuation)

stop_words = set(stopwords.words("english"))
remaining_words = []

for word in words:
    if word.lower() not in stop_words:
        remaining_words.append(word)

print(remaining_words)
```

```python
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
stemmed_words = []
for word in remaining_words:
    stemmed_words.append(stemmer.stem(word))

print(stemmed_words)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

lemmatized_words = []
for word in remaining_words:
    lemmatized_words.append(lemmatizer.lemmatize(word))


print(lemmatized_words)
```
